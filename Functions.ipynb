#Functions



1. What is the difference between a function and a method in Python?

  ->Function: A block of code defined using def or lambda that is not tied to an object.Example: def greet():

  Method: A function that is associated with an object (usually a class). It’s called using dot notation.Example: str.upper() is a method of string objects.
  
2. Explain the concept of function arguments and parameters in Python.

  ->Parameter: The variable listed in the function definition.Example: def greet(name): → name is a parameter.
  
  Argument: The actual value passed to the function when it is called.Example: greet("Alice") → "Alice" is an argument.

3. What are the different ways to define and call a function in Python?

  ->def add(a, b):  # Named function

    return a + b
    
    lambda_add = lambda a, b: a + b

4. What is the purpose of the return statement in a Python function?

  ->The return statement exits the function and sends a result back to the caller. Without return, the function returns None by default.

5. What are iterators in Python and how do they differ from iterables?

  ->Iterable: Any object you can loop over (like lists, tuples, strings).It implements __iter__() and returns an iterator.
  
  Iterator: An object that produces values one at a time using __next__() and remembers its state.

6. Explain the concept of generators in Python and how they are defined.

  ->A generator is a function that uses yield instead of return. It returns a generator object, which is an iterator.

  def count_up_to(n):i = 1 while i <= n:yield i i += 1

7. What are the advantages of using generators over regular functions?

  ->Memory efficient – doesn’t load all data at once.
  
  Lazy evaluation – computes values only when needed.
  
  Simple syntax for writing iterators

8. What is a lambda function in Python and when is it typically used?

  ->A lambda function is an anonymous, one-line function.
  
  Used with functions like map(), filter(), sorted().
  
  When you need a short, throwaway function.

9. Explain the purpose and usage of the map() function in Python.

  ->map() applies a function to each item in an iterable.

  nums = [1, 2, 3]
  squares = list(map(lambda x: x**2, nums))

10. What is the difference between map(), reduce(), and filter() functions in Python?
  
  ->Function	Purpose	Example
  
  map()	Transforms items	map(lambda x: x*2, [1, 2, 3]) → [2, 4, 6]
  
  filter()	Filters items	filter(lambda x: x % 2 == 0, [1, 2, 3]) → [2]
  
  reduce()	Reduces to a single value	reduce(lambda x, y: x + y, [1, 2, 3]) → 6 (Needs from functools import reduce)

```python
#Question 1
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
sum_of_even_numbers([1,2,3,4,5])
```

```python
#Question 4
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
is_prime(13)
```

```python
#Question 5
def fib(n):
  a=0
  b=1
  fibo=[]
  for i in range(n):
    fibo.append(a)
    a,b=b,a+b
  return fibo
lis=iter(fib(5))
print(next(lis))
print(next(lis))
print(next(lis))
print(next(lis))
print(next(lis))
```

```python
#Question 6
def power_of_two(n):
  num=2
  for i in range(n):
    yield num
    num=num*2
power=power_of_two(3)
print(next(power))
print(next(power))
print(next(power))
```

```python
#Question 7
def read_string_line_by_line(text):
    for line in text.splitlines():
        yield line
sample_text = """This is line one.
This is line two.
This is line three."""

for line in read_string_line_by_line(sample_text):
    print(line)
```
